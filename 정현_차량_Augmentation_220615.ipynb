```python
import albumentations as A
import cv2


transform = A.Compose([
    A.RandomCrop(width=256, height=256),
    A.HorizontalFlip(p=0.5),
    A.RandomBrightnessContrast(p=0.2),
    A.Rotate(limit=90, interpolation=1, border_mode=4, always_apply=False, p=0.5),
])
```
